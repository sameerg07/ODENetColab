```
import networkx as nx
import matplotlib.pyplot as plt
```

```
# x = networkx.generators.random_graphs.barabasi_albert_graph(4596,0.00248292738)
x = nx.generators.erdos_renyi_graph(9183,0.00358465455)
```

```
nx.draw(x, with_labels=False,node_size=10)
plt.show()
```

```
y = nx.generators.watts_strogatz_graph(9183,32,0.3)
```

```
nx.draw(y, with_labels=False,node_size = 10)
plt.show()
```

```
acc_x = nx.average_clustering(x)
acc_y = nx.average_clustering(y)
print(acc_x)
print(acc_y)
```

```
import random
final = []
#code creating G here
for (u,v,w) in y.edges(data=True):
    w['weight'] = random.randint(0,1)
    final.append(w['weight'])
```

```
labels = nx.get_edge_attributes(y,'weight')
nx.draw(y,with_labels=False,node_size = 10,edge_labels = labels)
plt.show()
```

```
y_dir = nx.to_directed(y)
triads_t = nx.triadic_census(y_dir)

#no of triads
print("The sum of each type of triads:")
triads_t
```

```
ratios_of_triads_t = []
sum_triads_t = sum(triads_t.values()) 

for i in triads_t.keys():
    ratios_of_triads_t.append((i,(triads_t[i]/sum_triads_t)))
    
print("The ratios of each type of traids:")
ratios_of_triads_t
```
